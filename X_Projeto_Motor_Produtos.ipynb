```python
import subprocess
import sys
import os
import webbrowser
import pathlib

BASE_DIR = pathlib.Path(os.getcwd())

print("=== MOTOR DE RECOMENDAÇÃO — SETUP E EXECUÇÃO ===\n")

# 1. Instala requirements
print("📦 Instalando dependências...")
result = subprocess.run([
    sys.executable, '-m', 'pip', 'install', '-r', 'requirements.txt', '-q'
], capture_output=False)
if result.returncode == 0:
    print("  ✓ Dependências instaladas!\n")
else:
    print("  ✗ Erro ao instalar dependências!")
    sys.exit(1)

# 2. Instala nbconvert se necessário
subprocess.run([sys.executable, '-m', 'pip', 'install', 'nbconvert', '-q'])

# 3. Executa notebooks
notebooks = [
    'notebooks/01_geracao_dados.ipynb',
    'notebooks/02_cluterizacao.ipynb',
    'notebooks/03_rfm_score.ipynb',
    'notebooks/04_score_propensao.ipynb',
    'notebooks/05_depara_segmentacao.ipynb',
    'notebooks/06_pipeline_incremental.ipynb',
    'notebooks/07_recomendacao.ipynb',
    'notebooks/08_produto_final.ipynb',
]

print("🚀 Executando notebooks...\n")
for nb in notebooks:
    print(f"  Rodando {nb}...")
    result = subprocess.run([
        sys.executable, '-m', 'nbconvert',
        '--to', 'notebook',
        '--execute',
        '--inplace',
        '--ExecutePreprocessor.timeout=1800',
        nb
    ], capture_output=True, text=True)

    if result.returncode == 0:
        print(f"  ✓ Concluído!\n")
    else:
        print(f"  ✗ ERRO em {nb}:")
        print(result.stderr[-500:])
        sys.exit(1)

# 4. Abre README como HTML
print("📄 Abrindo README...")
readme_path = BASE_DIR / 'README.md'
readme_html = BASE_DIR / 'docs' / 'readme_temp.html'

if readme_path.exists():
    with open(readme_path, 'r', encoding='utf-8') as f:
        md_content = f.read()

    html_readme = f"""<!DOCTYPE html>
<html lang="pt-BR">
<head>
<meta charset="UTF-8">
<title>Motor de Recomendação — README</title>
<style>
  body{{font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',sans-serif;
       max-width:900px;margin:40px auto;padding:0 20px;color:#1a1a2e;line-height:1.7}}
  h1{{color:#EC0000;border-bottom:3px solid #EC0000;padding-bottom:10px}}
  h2{{color:#1a1a2e;border-bottom:1px solid #eee;padding-bottom:6px}}
  h3{{color:#444}}
  code{{background:#f4f4f4;padding:2px 6px;border-radius:4px;font-size:13px}}
  pre{{background:#1a1a2e;color:#eee;padding:16px;border-radius:8px;overflow-x:auto}}
  pre code{{background:none;color:#eee}}
  a{{color:#EC0000}}
  table{{border-collapse:collapse;width:100%}}
  th{{background:#EC0000;color:white;padding:8px 12px}}
  td{{padding:8px 12px;border-bottom:1px solid #eee}}
  tr:hover td{{background:#fafafa}}
  blockquote{{border-left:4px solid #EC0000;margin:0;padding-left:16px;color:#666}}
  img{{max-width:100%;border-radius:8px}}
</style>
</head>
<body>
<pre id="md" style="display:none">{md_content}</pre>
<div id="content"></div>
<script src="https://cdn.jsdelivr.net/npm/marked/marked.min.js"></script>
<script>
  const md = document.getElementById('md').textContent;
  document.getElementById('content').innerHTML = marked.parse(md);
</script>
</body>
</html>"""

    with open(readme_html, 'w', encoding='utf-8') as f:
        f.write(html_readme)

    webbrowser.open(str(readme_html))
    print("  ✓ README aberto!\n")

# 5. Abre relatório final
relatorio = BASE_DIR / 'docs' / 'relatorio_final.html'
if relatorio.exists():
    print("📊 Abrindo relatório final...")
    webbrowser.open(str(relatorio))
    print("  ✓ Relatório aberto!\n")

# 6. Abre repositório GitHub
print("🐙 Abrindo repositório GitHub...")
webbrowser.open("https://github.com/calebe12/DATAMASTER_MOTOR_DE_RECOMENDACAO_PRODUTOS")
print("  ✓ Repositório aberto!\n")

print("=== TUDO CONCLUÍDO COM SUCESSO! ===")
```
